# **Calculated Fields, Parameters & LOD Expressions Assignment**
---

# **Question 1: What is a Calculated Field and why is it important?**

**Answer:**

A calculated field is a new field created using formulas based on existing data in Tableau.
It allows us to perform calculations such as profit, ratios, or conditions.

**Importance:**

* Helps in creating new insights from raw data
* Used to calculate metrics like Profit, Profit Ratio, etc.
* Makes dashboards more meaningful and analytical
* Reduces dependency on source data

**In simple terms:**

It helps us **convert raw data into useful information**

---

# **Question 2: Why can’t parameters change visuals directly in Tableau?**

**Answer:**

Parameters cannot change visuals directly because they are not connected to the data by default.
They only store a value and do not perform any action unless used in a calculated field.

To make parameters affect visuals, we must use them inside calculations, filters, or logic.

**In simple terms:**

Parameters are just inputs — they need calculations to work on visuals.

---

# **Question 3: When should EXCLUDE LOD be used?**

**Answer:**

EXCLUDE LOD is used when we want to remove a dimension from the current view and calculate values at a higher level.

It helps in comparing detailed data with overall or higher-level data.

**Example:**

* Removing Category to calculate total sales across all categories

**In simple terms:**

Use EXCLUDE when you want to **ignore a dimension in calculation**

---

# **Question 4: Difference between FIXED and INCLUDE LOD**

**Answer:**

**FIXED LOD:**

* Calculates values at a fixed level of detail
* Ignores most filters (like Category)
* Example: Sales per City regardless of view

**INCLUDE LOD:**

* Adds extra detail to the current view
* Works with existing dimensions
* Example: Product-level calculation even when viewing Category

### **Key Difference:**

| FIXED                   | INCLUDE               |
| ----------------------- | --------------------- |
| Ignores view level      | Works with view level |
| Fixed calculation level | Adds extra detail     |
| Independent of filters  | Depends on view       |

---

**Simple line to remember:**

* FIXED → Fixes the level
* INCLUDE → Adds more detail
